```python
import os
import csv

import matplotlib.pyplot as plt

import source.visualization as vis
import source.config as config
```

```python
results = vis.load_results(config.RESULTS_DIRECTORY)
```

Saving results to csv file format

```python
csv_folder = os.path.join(".data", "csv")

for label in results:
    with open(os.path.join(csv_folder, label + ".csv"), 'w') as file:
        writer = csv.writer(file, lineterminator='\n')
        writer.writerow(
            [f"Iteration {i}" for i in range(len(results[label]))]
        )
        for i in range(len(results[label][0])):
            writer.writerow(
                [iteration[i][1] for iteration in results[label]]
            )
```

Best/Average/Worst diagrams

```python
for label in results:
    best_history, average_history, worst_history = vis.gather_all_statistics(results[label])
    
    vis.plot_diagrams(best_history, average_history, worst_history, label)
```

Histograms for final population

```python
for label in results:
    population = results[label][-1]
    vis.plot_histogram([specimen[1] for specimen in population], label)
```
