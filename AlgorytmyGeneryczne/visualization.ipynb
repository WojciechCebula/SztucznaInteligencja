```python
import matplotlib.pyplot as plt

import source.visualization as vis
import source.config as config
```

```python
results = vis.load_results(config.RESULTS_DIRECTORY)
```

```python
for label in results:
    best_history, average_history, worst_history = vis.gather_all_statistics(results[label])
    
    fig = plt.figure()
    fig.tight_layout()
    ax = fig.add_subplot()
    ax.set_title(label)
    ax.set_xlabel("Iteration")
    ax.set_ylabel("Score")
    
    ymax = worst_history[0]
    xmax = len(worst_history)
    
    ax.plot(worst_history, color="r", label="worst")
    ax.plot(average_history, color="y", label="average")
    ax.plot(best_history, color="g", label="best")
    
    vis.add_annotation(ax, worst_history, ymax, xmax)
    vis.add_annotation(ax, average_history, ymax, xmax)
    vis.add_annotation(ax, best_history, ymax, xmax)
    ax.legend()

    plt.show()

```
